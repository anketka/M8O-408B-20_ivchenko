```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
a, b, n = 0, 1, 51
h = (b - a) / (n - 1)
```

```python
def read_txt_file(filename):
    with open(filename, 'r') as f:
        matrix = [np.array(list(map(float, line.split()))) for line in f]
    return np.array(matrix)


analytical_solution = read_txt_file("analytical_solution.txt")[0]
quadrature_trapezoid_method = read_txt_file("quadrature_trapezoid.txt")[0]
quadrature_Simpson_method = read_txt_file("quadrature_Simpson.txt")[0]
iterations_trapezoid_method = read_txt_file("iterations_trapezoid.txt")[0]
iterations_Simpson_method = read_txt_file("iterations_Simpson.txt")[0]

solutions = {
    'Аналитическое решение': analytical_solution,
    'Метод квадратур с использованием метода трапеций': quadrature_trapezoid_method,
    'Метод квадратур с использованием метода Симпсона': quadrature_Simpson_method,
    'Метод простых итераций с использованием метода трапеций': iterations_trapezoid_method,
    'Метод простых итераций с использованием метода Симпсона': iterations_Simpson_method
}
```

```python
def plot_solutions(solutions, a, b, h):
    x = np.arange(a, b + h, h)
    plt.figure(figsize=(15, 9))
    for method_name, solution in solutions.items():
        plt.plot(x, solution, label=method_name)

    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_solutions(solutions, a, b, h)
```

```python
def abs_error(A, B):
    return abs(A - B)
```

```python
def plot_errors(solutions, a, b, h):
    x = np.arange(a, b + h, h)
    plt.figure(figsize=(15, 9))
    for method_name, solution in solutions.items():
        plt.plot(x, abs_error(solutions['Аналитическое решение'], solution), label=method_name)

    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_errors(solutions, a, b, h)
```
